```python
import matplotlib.pyplot as plt
from functools import reduce
```

```python
x1, y1 = reduce(lambda x,y : [x[0] + [y.split(",")[0]], x[1] + [int(y.split(",")[1])]], pythonFile.split("\n"),[[],[]])
```

```python
x2, y2 = reduce(lambda x,y : [x[0] + [y.split(",")[0]], x[1] + [int(y.split(",")[1])]], ipynbFile.split("\n"),[[],[]])
```

```python
frequency = dict()
for i, function in enumerate(x1):
    val = frequency.get(function, [0,0])
    val[0] = val[0] + y1[i]
    frequency[function] = val
for j, function in enumerate(x2):
    val = frequency.get(function, [0,0])
    val[1] = val[1] + y2[j]
    frequency[function] = val
```

```python
def getTopK(frequency, k):
    return sorted(frequency.items(), key=lambda x: x[1][0])[-k:][::-1]
```

```python
getTopK(frequency,20)
```

```python
def plotScatterForPyAndIpynb(k = 20):
    fig, ax1 = plt.subplots()
    handle1 = ax1.scatter(range(k), list(map(lambda x: x[1][0], getTopK(frequency,k))), c='b')
    ax1.tick_params('y', colors='b')
    plt.xticks(rotation=90)
    plt.xticks(range(k), list(map(lambda x: x[0], getTopK(frequency,k))))
    ax2 = ax1.twinx()
    handle2 = ax2.scatter(range(k), list(map(lambda x: x[1][1], getTopK(frequency,k))), c='r')
    ax2.tick_params('y',colors='r')
    plt.legend([handle1, handle2],['python file', 'ipynb'])
    plt.show()

plotScatterForPyAndIpynb()
```
